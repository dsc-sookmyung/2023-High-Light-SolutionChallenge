```python
import fitz
import PIL.Image
import io
import re
from pdfminer.high_level import extract_pages, extract_text

# for page_layout in extract_pages("temp_dataset/os.pdf"):
#     for element in page_layout:
#         print(element)
#     print()

# text = extract_text('temp_dataset/os.pdf')
# print(text)

pdf = fitz.open('temp_dataset/os.pdf')
counter = 1

# text -> image 순으로 페이지당으로 나옴
for i in range(len(pdf)):
    page = pdf[i] # load page
    images = page.get_images()
    text = page.get_text('sentences')
    print(text)
    for image in images:
        base_img = pdf.extract_image(image[0])
        image_data = base_img['image']
        img = PIL.Image.open(io.BytesIO(image_data))
        extension = base_img['ext']
        img.save(open(f"temp_dataset/split_pages/image{counter}.{extension}", "wb"))
        counter += 1
        
```
